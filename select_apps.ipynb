```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import dataset to identify top apps in terms of number of downloads
df = pd.read_csv('04-data/Google-Playstore.csv')
```

```python
df1 = df[df['Developer Id'] != 'Google LLC']
df1.shape
```

```python
test = df1[(df1['Maximum Installs'] <= 10000000) & (df1['Maximum Installs'] >= 500000)]
test.shape
```

```python
test1 = test.loc[test['App Name'].str.contains(r'[^\x00-\x7F]+') == False]
```

```python
test1.sort_values('Maximum Installs',ascending=False).tail(60)
```

```python
test1.groupby('Ad Supported').count()
```

```python
plt.figure(figsize=[12,4])
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.hist(test1['Maximum Installs'])
plt.title('Max installs histogram',family='arial',fontsize=14)
plt.tight_layout()
plt.xlabel('Installs')
plt.ylabel('Number')

plt.tight_layout()
plt.show;
```

```python
test1.to_csv('04-data/filtered_data.csv')
```
